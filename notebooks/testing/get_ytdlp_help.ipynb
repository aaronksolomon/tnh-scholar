```python
import sys
from io import StringIO
from pathlib import Path

import yt_dlp

from tnh_scholar.utils.file_utils import write_text_to_file
```

```python


# Temporarily redirect stdout to capture the help text
old_stdout = sys.stdout
help_output = StringIO()
sys.stdout = help_output

# Get the help text
help(yt_dlp.YoutubeDL)

# Restore stdout and get the string
sys.stdout = old_stdout
help_text = help_output.getvalue()
help_output.close()
```

```python
write_text_to_file(Path("yt_dlp_help_output"), help_text, overwrite=True)
```
